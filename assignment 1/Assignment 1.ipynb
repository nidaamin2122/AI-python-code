```python
sub1 = int(input("Enter Marks for English: "))
sub2 = int(input("Enter Marks for Maths: "))
sub3 = int(input("Enter Marks for Science: "))
sub4 = int(input("Enter Marks for Urdu: "))
sub5 = int(input("Enter Marks for Islamiat: "))

total_marks = sub1 + sub2 + sub3 + sub4 + sub5
percentage = (total_marks / 500) * 100

print("    Mark Sheet          ")
print("Total Marks:", total_marks)
print("Percentage:", percentage, "%")

if percentage >= 80:
    print("Grade: A-One")
elif percentage >= 70:
    print("Grade: A")
elif percentage >= 60:
    print("Grade: B")
elif percentage >= 50:
    print("Grade: C")
elif percentage >= 40:
    print("Grade: D")
else:
    print("Grade: Fail")
```
